# Problem Statement

This anlaysis study is counducted to find the association between items as a  basket analysis, in order to sell more products and gain profit.

In this study case, we will use apriori algorithm in python to find the association rules between items.

# Importing the required libraries

```
import pandas as pd 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import io
```

# Loading The Datasdet

```
from google.colab import files
uploaded = files.upload()
```

```
df = pd.read_csv(io.BytesIO(uploaded['OnlineRetail.csv']),  encoding= 'unicode_escape')
print(df.shape)
print(df.describe())
print(df.info())
print(df.head(10))
```

# Preprocessing the data

##Handling missing values

```
print(df.columns[df.isnull().any()])
```

```
df.dropna(axis=0, how='any', subset=['Description'], inplace=True)
```

```
df
```

```
print(df.columns[df.isnull().any()])
```

## Removing Outliers

As Unit Price and quantity can't be negative we are removing those rows contaning them.

```
df1 = df[df['Quantity'] > 0] 
```

```
df1 = df1[df1['UnitPrice'] > 0] 
```

```
print(df1.describe())
```

## Shaping the dataset 

```
df1['Country'].describe()
```

```
australiadf= (df1[df1['Country']=='Australia'].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))

australiadf.head()
```

Making sure that any product bought more than 1 the value will be 1 and if less than 0 then 0.

```
def encoding_values(x):
  if x<=0:
    return 0
  if x>=1:
    return 1
australiadf_basket=australiadf.applymap(encoding_values)
australiadf_basket.drop('POSTAGE',inplace=True, axis=1)
australiadf_basket
```

#Finding the association rules

Finding the frequent items where 8% or more of customers bought them together.

```
frequent_items=apriori(australiadf_basket, min_support=0.07,use_colnames=True)
rules=association_rules(frequent_items,metric="lift",min_threshold=1)
rules.head()
```

```
rules[(rules['lift']>=10) & (rules['confidence']>0.8)]
```

#Results

**The first rule:**

With **100%** confidence that **7%** of customers who purchased (RED RETROSPOT CAKE STAND) also purchased (PENCILS TUBE RED RETROSPOT).

And the expectations of a customer buying (PENCILS TUBE RED RETROSPOT) when they bought (RED RETROSPOT CAKE STAND) is **14.25** times increased.

**The fourth rule:**

With **100%** confidence that **7%** of customers who purchased (SET OF 4 PANTRY JELLY MOULDS) also purchased (FELTCRAFT DOLL MOLLY).

And the expectations of a customer buying (FELTCRAFT DOLL MOLLY) when they bought (SET OF 4 PANTRY JELLY MOULDS) is **14.25** times increased.
